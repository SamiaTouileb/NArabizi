```python
train = train.drop_duplicates(ignore_index=True)
```

There are reference numbers in some examples

```python
joined = train["sentiment"].values.astype(str) + np.array([" | "] * train.shape[0], dtype=object) + train["review"].values
```

```python
re.findall(r".*\|.*\d+.*", "\n".join(joined))
```

Drop duplicates and remove references over all datasets and export

```python
path = "../data/sentiment/bg/"

for filename in os.listdir(path):
    table = pd.read_csv(path + filename, header=None)
    table.columns = ["sentiment", "review"]
    table = table.drop_duplicates(ignore_index=True)
    table["review"] = table["review"].apply(lambda x: re.sub(r"#?\d{5,}[/\d+]*", "", x))
    table = table[table["review"].str.len() > 0]
    table.to_csv(path + filename, index=False, header=False)
```
